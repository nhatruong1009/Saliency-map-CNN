```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

def get_file_paths(folder_path):
    files = os.listdir(folder_path)
    return files

def visualize(path, num_size):
    # Load the image
    image = cv2.imread(path)
    # Define the number of rows and columns for the grid
    rows = num_size
    cols = num_size

    # Calculate the spacing between grid lines
    spacing_x = image.shape[1] // (cols + 1)
    spacing_y = image.shape[0] // (rows + 1)

    # Draw vertical grid lines
    for i in range(1, cols + 1):
        x = i * spacing_x
        if i % 4 == 1:
            cv2.line(image, (x, 0), (x, image.shape[0]), (0, 255, 0), 1)
        elif i % 4 == 2:
            cv2.line(image, (x, 0), (x, image.shape[0]), (255, 0, 0), 1)
        elif i % 4 == 3:
            cv2.line(image, (x, 0), (x, image.shape[0]), (255, 0, 255), 1)
        else:
            cv2.line(image, (x, 0), (x, image.shape[0]), (255, 255, 0), 1)

    # Draw horizontal grid lines
    for i in range(1, rows + 1):
        y = i * spacing_y
        
        if i % 4 == 1:
            cv2.line(image, (0, y), (image.shape[1], y), (0, 255, 0), 1)
        elif i % 4 == 2:
            cv2.line(image, (0, y), (image.shape[1], y), (255, 0, 0), 1)
        elif i % 4 == 3:
            cv2.line(image, (0, y), (image.shape[1], y), (255, 0, 255), 1)
        else:
            cv2.line(image, (0, y), (image.shape[1], y), (255, 255, 0), 1)

    # Convert BGR to RGB for matplotlib
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Display the image with grid using matplotlib
    plt.imshow(image_rgb)
    plt.axis('off')  # Turn off axis
    plt.show()
```

```python
k = 0
# Đường dẫn đến thư mục chứa các tệp
folder_path = '../../ILSVRC2012_val_folders'
# folder_path = 'images_test'
file_paths = get_file_paths(folder_path)

image_path = f"{folder_path}/{file_paths[k]}"
```

```python
visualize(image_path, 6)
```

```python
visualize(image_path, 13)
```

```python
visualize(image_path, 27)
```
